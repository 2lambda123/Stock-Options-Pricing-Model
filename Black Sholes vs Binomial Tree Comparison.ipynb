__European Option Pricing Comparison Between Binomial Tree and Black Sholes Merton Model__

```python
from BS_euro_option import *
from stock_option import stockoption
from euro_option import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Comparison = pd.DataFrame(columns=["Black Sholes Model European Option Price","Binomial Tree Model European Option Price",
                                   "Time Steps","Difference","Percentage Difference"])
for i in range(5,200,5):
    stockopt = stockoption(217.58, 215, 0.05, 0.1,i,'Log',{'tk': 'AAPL', 'is_calc': True, 'start': '2017-08-18',
                                                     'end': '2018-08-18', 'eu_option':False})
    option_eu = euro_option(stockopt)
    bt_price = option_eu.begin_tree(stockopt)
    bs_price = BS_euro_option(217.58, 215, 0.05, 0.1, 150,'Log',{'tk': 'AAPL', 'is_calc': True, 'start': '2017-08-18',
                                                     'end': '2018-08-18', 'eu_option':False})
    bs_price = bs_price.euro_vanilla_call_incl_div(stockopt)
    Comparison.loc[(i/5-1),"Black Sholes Model European Option Price"] = bs_price
    Comparison.loc[(i/5-1),"Time Steps"] = i
    Comparison.loc[(i/5-1),"Binomial Tree Model European Option Price"] = bt_price[0,0]
    Comparison.loc[(i/5-1),"Difference"] = bs_price - bt_price[0,0]
    Comparison.loc[(i/5-1),"Percentage Difference"] = ((bs_price - bt_price[0,0])/bs_price)*100
```

```python
Comparison
```

```python
fig,ax = plt.subplots(figsize=(14,8))
Comparison.plot(x="Time Steps",y="Binomial Tree Model European Option Price",ax=ax)
Comparison.plot(x="Time Steps",y="Black Sholes Model European Option Price",ax=ax)
plt.show()
```
